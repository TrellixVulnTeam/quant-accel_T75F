### Implied Timescales

```python
def it_helper(args):
    lt, trajsl = args
    from mixtape.markovstatemodel import MarkovStateModel
    tmsm = MarkovStateModel(lag_time=lt, n_timescales=10, reversible_type='mle')
    tmsm.fit(trajsl)
    return lt, tmsm.timescales_
```

```python
from IPython.parallel import Client
from itertools import repeat
```

```python
client = Client()
```

```python
lbv = client.load_balanced_view()
lbv.block = True
```

```python
lagtimes = range(1, 60, 2)
its = lbv.map(it_helper, zip(lagtimes, repeat(trajsl)))
```

```python
for lt, ts in its:
    plt.scatter(np.repeat(lt, len(ts)), ts)
plt.yscale('log')
```
