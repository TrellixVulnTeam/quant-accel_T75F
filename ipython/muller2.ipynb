# Load

```python
def load(which):
    walkypath = '/home/harrigan/biox/projects/quant-accel/muller4/'
    results = list()
    with open(os.path.join(walkypath, "%s-filelist.dat" % which)) as f:
        for line in f:
            line = line.strip()
            if len(line) == 0: continue
            with open(os.path.join(walkypath, line)) as pf:
                res = pickle.load(pf)
                rem = re.search(r"runcopy-([0-9]+)", line)
                if rem is not None:
                    res.params['runcopy'] = int(rem.group(1))
                results.append(res)
    return results
    
```

# Speedup

```python
import collections
```

```python
poplts = collections.defaultdict(lambda:np.inf)
for r in popresults:
    if r.params['lt'] == '1':
        poplts[r.params['tpr']] = min(r.poptime, poplts[r.params['tpr']])
```

```python
def calc_speedup(usespeedup, usetime=True):
    atspt = collections.defaultdict(lambda: np.zeros((0,2)))
    attpr = collections.defaultdict(lambda: np.zeros((0,2)))
    for r in popresults:
        
        if usetime and usespeedup:
            time = r.poptime
            speed =  poplts[r.params['tpr']] / time
        elif usetime and not usespeedup:
            time = r.poptime
            speed = poplts['1']/time
        elif not usetime and not usespeedup:         
            speed = r.popind + 1
        else:
            print "Not supported"
        
            
            
        if r.params['lt'] != '1':            
            atspt[r.params['spt']] = np.append(atspt[r.params['spt']], [[int(r.params['tpr']), speed]], axis=0)
            attpr[r.params['tpr']] = np.append(attpr[r.params['tpr']], [[int(r.params['spt']), speed]], axis=0)

    print "spts", atspt.keys()
    print "tprs", attpr.keys()
    return atspt, attpr
```

```python
def avg_and_errorbar(v):
    vxs = sorted(set(v[:,0]))
    w = np.zeros((len(vxs), 3))
    for i, vx in enumerate(vxs):
        selex = np.where(v[:,0]==vx)[0]
        w[i, 0] = vx
        w[i, 1] = np.exp(np.mean(np.log(v[selex, 1])))
        w[i, 2] = np.std(v[selex, 1])
    return w
```

```python
atspt, attpr = calc_speedup(usespeedup=False)
#popltsspeed = np.asarray(poplts.items(), dtype=float)
#popltsspeed[:,1] = poplts['1'] / popltsspeed[:,1]
#ites = atspt.items() + [(-1, popltsspeed)]
ites = atspt.items()
for k, v in sorted(ites, key=lambda x: -int(x[0])):
    w = avg_and_errorbar(v)
    #pp.plot(v[:,0], v[:,1], 'o', label=k)
    pp.errorbar(w[:,0], w[:,1], yerr=w[:,2], fmt='o-', label=k)
        
pp.xscale('log')
pp.xlabel('tpr')
pp.ylabel('speed')
pp.legend(loc='lower right', bbox_to_anchor=((1.5,0.0)))
pp.axhline(1, c='grey')
xmin, xmax = pp.xlim()
pp.xlim(min(10**-0.5, xmin), max(10**3.5, xmax))
ymin, ymax = pp.ylim()
#pp.ylim(min(-0.1, ymin), ymax)
#pp.yscale('symlog')
pp.show()
```

## Hists

```python
def hists(v):
    ally = np.log(v[:,1])
    allx = v[:,0]
    vxs = sorted(set(allx))
    w = np.zeros((len(vxs), 3))
    for i, vx in enumerate(vxs):
        selex = np.where(allx==vx)[0]

        y = ally[selex]
        
        # make hists
        #pp.plot(np.repeat([vx], len(selex)), v[selex,1], 'o', label=vx)
        hist, bin_edges = np.histogram(y, bins=10, range=((np.min(ally)-1, np.max(ally)+1)))
        bincs = np.diff(bin_edges)/2.0 + bin_edges[:-1]
        pp.plot(bincs, hist, 'o-', label=int(vx))
        pp.title(vx)
        pp.show()
        
```
