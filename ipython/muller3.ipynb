#Convergence over time
## Population
We visualize convergence (as defined by equilibrium population total variational distance (TVD)) over time. Each colored line is a a certain degree of parallelization (tpr) and each plot is at a certain round length (spt)

```python
uniques = muller.get_unique('pop', ('lt', 'spt'))
uniques = [(ss, ax) for ax, (sv, ss) in enumerate(uniques)]
uniqued = dict(uniques)
fig, axes = plt.subplots(3, 3, figsize=(16,12))
for r in muller.get_runcopy(7, 'pop'):
    paramtuple = (r.params['lt'], r.params['spt'])
    figname = "lt-%s_spt-%s" % paramtuple
    ax = axes[np.unravel_index(uniqued[figname], axes.shape)]
    ax.plot(r.errors[:,0], r.errors[:,1], '.-', label=r.params['tpr'])
    ax.set_title("LT %s, SPT %s" % paramtuple)
    ax.set_xlabel('walltime'); ax.set_ylabel('eq pop TVD')
    ax.axhline(0.6, c='r')
    ax.set_xscale('log')
```

## Histograms
We show that the number of rounds is non-skew distributed on a log scale

```python
fig, axes = plt.subplots(8, 5, figsize=(16, 4*8))
for i, (k, v) in enumerate(rounds_vs_spt.items()):
    for j, (hist, bin_centers, xval) in enumerate(sc.histogram(v, n_bins=10, whole_range=False, log=False)):
        ax = axes[j, i]
        ax.plot(bin_centers, hist, 'o-')
        ax.set_xlabel('# of rounds');
        ax.set_title("tpr: %s spt: %s" % (k, xval))
```

### Fit things

```python
rounds_vs_spt = muller.plot_vs('pop', 'rounds', xaxis='spt', label='tpr')
# One traj normalization
one_traj_fit = sc.FitResult(1, rounds_vs_spt[1])
one_traj_fit.fit_vs_spt(1.0)
rounds_vs_spt.otf = one_traj_fit
sfig, sax = pp.subplots()
tfig, tax = pp.subplots()
for fit in rounds_vs_spt.enum_fit():
    cfig, (ax1, ax2) = pp.subplots(1,2, figsize=(16,5))
    
    # Scatter points
    ax1.scatter(fit.x, fit.y)
    
    # Plot fit
    ax1.plot(fit.fitx, fit.fity, 'b-')    
    
    # Plot non-adaptive
    ax1.plot(fit.fitx, fit.na_scale, 'r-', label='Non-adaptive')
    
    # Plot Rounds = 1
    ax1.axhline(1.0, c='grey')
    
    # Plot speed on right and on top fig
    ax2.plot(fit.fitx, fit.fit_time, '-')
    ax2.scatter(fit.x, fit.dat_time)
    
    sax.plot(fit.fitx, fit.speedup, '-', label=fit.key)
    tax.plot(fit.fitx, fit.speed, '-', label=fit.key)
    
    # Format
    ax1.set_xscale('log'); ax1.set_yscale('log')
    ax1.set_xlabel('Steps per traj'); ax1.set_ylabel('Rounds')
    ax1.set_title('Tpr %s' % fit.key)
    
    ax2.set_xlabel('spt'); ax2.set_ylabel('time')
    ax2.set_xscale('log'); #ax2.set_yscale('log')

sax.set_xlabel('spt'); sax.set_ylabel('speedup')
sax.set_xscale('log'); sax.set_yscale('log')
sax.legend(loc='lower right', bbox_to_anchor=((1.5,0.0)))

tax.set_xlabel('spt'); tax.set_ylabel('speedup to tpr=1')
tax.set_xscale('log'); tax.set_yscale('log')
tax.legend(loc='lower right', bbox_to_anchor=((1.5,0.0)))
```
