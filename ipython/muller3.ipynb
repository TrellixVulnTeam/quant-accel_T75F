# Fancy selection of last SPT for calculation of non-adaptive
We find the spt of interest and then construct a function which we pass in that just selects those at the last spt

```python
# Find SPT's of interest
maxspts = dict()
for k, v in rounds_vs_spt.items():
    maxspts[k] = np.max(v[:,0])
    print "tpr", k, "has max spt =", np.max(v[:,0])
```

```python
# Define our function
def where_func(r):
    # Why not hack in the following while we're at it?
    r.params['adaptive'] = False
    return r.params['spt'] == maxspts[r.params['tpr']]
```

```python
# Get our dicts
time_vs_spt_na = muller.plot_vs('sub', 'time', xaxis='spt', label='tpr', where=where_func)
time_vs_tpr_na = muller.plot_vs('sub', 'time', xaxis='tpr', label='adaptive', where=where_func)
```

```python
# See if it worked
for k, v in time_vs_spt_na.items():
    print k
    print v[:,0].shape
```
