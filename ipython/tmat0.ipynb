```python
def plot_its(comps, labels, i, gold_its):
    for comp, lab in zip(comps, labels):
        pp.plot(comp[:,0], comp[:,i], 'o', label=lab)
    pp.yscale('log')
    xmin, xmax = pp.xlim()
    pp.hlines(gold_its[i-1], xmin, xmax)
    pp.legend()
    pp.show()
def plot_errs(comps, labels, i, gold_its):
    for comp, lab in zip(comps, labels):
        errs = (np.log(comp[:,i]/gold_its[i-1]))**2
        pp.plot(comp[:,0], errs, 'o', label=lab)
    pp.legend(loc='upper right')
```
