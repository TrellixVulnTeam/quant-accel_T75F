```python
plt.subplots(figsize=(9,5))
ref = msm.eigenvectors_[:,1]
est = np.copy(msm.eigenvectors_[:,1]) + np.random.normal(0,0.4)
scale = 2e3

# positive
plt.scatter(cents[ref>=0,0], cents[ref>=0,1], linewidths=0, s=scale*ref[ref>=0],
            c='coral')

# Neg
plt.scatter(cents[ref<0,0], cents[ref<0,1], linewidths=0, s=-scale*ref[ref<0],
            c='skyblue')

plt.scatter(cents[est>=0,0], cents[est>=0,1], s=scale * est[est>=0],
            marker='o', facecolors='none', edgecolors='r', linewidths=2)

plt.scatter(cents[est<0,0], cents[est<0,1], s=-scale * est[est<0],
            marker='o', facecolors='none', edgecolors='b', linewidths=2)
```
